```python
#eliminaçao de colunas com muitos dados faltantes ou dados nao numericos, ou dados irrelevantes
dados.drop('q8.20', axis=1, inplace=True)
dados.drop('q8.18', axis=1, inplace=True)
dados.drop('q8.17', axis=1, inplace=True)
dados.drop('q8.10', axis=1, inplace=True)
dados.drop('q8.9', axis=1, inplace=True)
dados.drop('q8.8', axis=1, inplace=True)
dados.drop('q8.2', axis=1, inplace=True)
dados.drop('q1_1.personal.opinion.of.this.Deodorant', axis=1, inplace=True)
dados.drop('Product', axis=1, inplace=True)
dados.drop('Respondent.ID', axis=1, inplace=True)
```

```python
#dados.isnull().sum()/len(dados['Product.ID'])*100
```

```python
#preenchimento de valores faltantes com valores medios ou mediana
dados['q8.12'].fillna(dados['q8.12'].median(), inplace=True)
dados['q8.7'].fillna(dados['q8.7'].median(), inplace=True)
```

```python
dadosx = dados.drop('Instant.Liking', axis=1) 
dadosy = dados['Instant.Liking']
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
#Criando a lista de parametros a serem otimizados
#valores_C = np.array([0.01, 0.1, 0.5, 1, 2, 3, 5, 10, 20, 50, 100])
#regularizacao = ['l1', 'l2']
#valoresgrid = {'C':valores_C , 'penalty': regularizacao}
valoresgrid = {'C': [0.01, 0.1, 0.5, 1, 2, 3, 5, 10, 20, 50, 100], 'penalty': ['l1', 'l2']}
#criando o modelo
modelo = LogisticRegression(solver='liblinear')
resultado = GridSearchCV(estimator = modelo, param_grid = valoresgrid, cv=5)
resultado.fit(dadosx, dadosy)

print('Melhor score (coef R2):', resultado.best_score_)
print('Melhor valor do parametro C:', resultado.best_estimator_.C)
print('Melhor regularizador (l1 ou l2):', resultado.best_estimator_.penalty)
#2022.04.26
print('Best Parameters:',resultado.best_params_)
print('Coeficientes da regressao - pesos das variaveis independentes(x):\n',resultado.best_estimator_.coef_)
print('Constante da regressao:',resultado.best_estimator_.intercept_)
df = pd.DataFrame(resultado.cv_results_)
df.to_csv('cv_results.csv')
#print('cv results:', resultado.cv_results_)
```
