```python
#Os 6 valores que representam o estado do robô consistem no seno e cosseno dos angulos e velocidades
#angulares dos braços roboticos.
#As 3 ações consistem em aplicar um torque de +1, 0 ou -1 na junção movel entre os dois braços.
#O objetivo é fazer o braço do robo tocar na linha superior da figura
# A cada timestep, a recompensa obtida é sempre -1, exceto se o objetivo é antingido, recebendo
# zero neste caso.
```

```python
import time
env=gym.make('Acrobot-v1')
obs=env.reset() #ate aqui temos somente o estado inicial do ambiente definido
lista_recompensas=[]
for i in range(1000):
    env.render()
    time.sleep(0.1)
    acao=env.action_space.sample() #gera uma nova ação aleatoria
    print(obs)
    obs, recompensa, terminou, info, = env.step(acao) #a partir de cada ação, vai para um novo
    # estado, recebe recompensa, diz se o episodio terminou (done) e traz um dicionario com 
    # outras informações, se houver.
    lista_recompensas.append(recompensa)
    if terminou: #se terminou um episodio
        print('Episodio finalizado depois de {} timesteps'.format(i+1))
        break
env.close() # fecha a janela
```

```python
env=gym.make('Acrobot-v1')

modelo = DQN(policy=MlpPolicy, env=env, gamma=0.99, learning_rate=0.0005, buffer_size=5000, 
            exploration_fraction=0.1, exploration_final_eps=0.1, exploration_initial_eps=1.0, 
            batch_size=32, double_q=False, learning_starts=5000, verbose=1)
modelo.learn(total_timesteps=50000, log_interval=10)
```

```python
import time
# Visualizando o modelo treinado:
obs = env.reset()
terminou = False
recompensa_total = 0
for i in range(5):
    while terminou!= True:
        acao, estado = modelo.predict(obs)
        obs, recompensa, terminou, info = env.step(acao)
        recompensa_total+=recompensa
        env.render()
        time.sleep(0.05) 
    print('Recompensa total:', recompensa_total)
    obs=env.reset()
    recompensa_total=0
    terminou=False
    time.sleep(1)
env.close()
```
