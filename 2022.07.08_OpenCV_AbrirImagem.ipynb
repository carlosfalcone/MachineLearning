## Rotacionando uma imagem

```python
#Obtendo as dimensoes da imagem e seu centro
(altura,largura)=imagem_rgb.shape[:2]
centro=(largura//2,altura//2)
#Rotacionando a imagem
param_rot=cv2.getRotationMatrix2D(centro,45,1.0) #(centro, angulo_rotaçao, escala_zoom)
rotacionado=cv2.warpAffine(imagem_rgb, param_rot,(largura,altura)) #(imagem, parametros,dimensoes_imagem_saida)
plt.imshow(rotacionado)
```

## Deslocando uma imagem

```python
#Obtendo as dimensoes da imagem e seu centro
(altura,largura)=imagem_rgb.shape[:2]
#Deslocando a imagem
param_shift=np.float32([[1,0,-46],[0,1,120]]) #deslocamento pra esquerda de -46 e para baixo em 120
deslocado=cv2.warpAffine(imagem_rgb, param_shift,(largura,altura)) #(imagem, parametros,dimensoes_imagem_saida)
plt.imshow(deslocado)
```

## Redimensionando uma imagem

```python
#Obtendo as dimensoes da imagem
(altura,largura)=imagem_rgb.shape[:2]
nova_largura=180
#Mantendo a proporçao da imagem
proporcao=altura/largura
nova_dim=(nova_largura,int(nova_largura*proporcao)) #nova_largura*proporcao=nova_altura
#Executando o redimensionamento da imagem
redim=cv2.resize(imagem_rgb,nova_dim,interpolation=cv2.INTER_AREA) #INTER_AREA, INTER_LINEAR
plt.imshow(redim)
```

## Criando histogramas de imagens

## Aplicando operaçoes morfologicas em imagens

```python
imagem_cerebro=cv2.imread('C:/Users/Falcone/Documents/0_MachineLearning/2022.07.08_Envirom_OpenCV/Brain_SaltPepperNoise.jpg')
```

```python
imagem_GoldenGate=cv2.imread('C:/Users/Falcone/Documents/0_MachineLearning/2022.07.08_Envirom_OpenCV/dicas-de-las-vegas-golden-gate_baixaresol.jpg') 
imagem_GoldenGateRGB=cv2.cvtColor(imagem_GoldenGate,cv2.COLOR_RGB2BGR)
imagem_GoldenGateCinza=cv2.cvtColor(imagem_GoldenGate,cv2.COLOR_BGR2GRAY)
```

```python
#Erosao: kernel central assume o menor valor dos demais pixels (Preto)
imagem_GoldenGateCinza
imagem_erodida=cv2.erode(imagem_GoldenGateCinza.copy(),(5,5),iterations=3)
```

```python
cv2.imshow('Imagem Original',imagem_GoldenGateCinza)
cv2.imshow('Imagem Erodida',imagem_erodida)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Dilataçao: kernel central assume o MAIOR valor dos demais pixels (Branco)
imagem_GoldenGateCinza
imagem_dilatada=cv2.dilate(imagem_GoldenGateCinza.copy(),(5,5),iterations=50)
```

```python
cv2.imshow('Imagem Original',imagem_GoldenGateCinza)
cv2.imshow('Imagem Erodida',imagem_dilatada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Opening: Erosao e depois Dilataçao (tira o ruido branco - Salt)
imagem_cerebro_erodida=cv2.erode(imagem_cerebro.copy(),(5,5),iterations=5)
imagem_cerebro_erodida_dilatada=cv2.dilate(imagem_cerebro_erodida.copy(),(5,5),iterations=5)
cv2.imshow('Imagem Original',imagem_cerebro)
cv2.imshow('Imagem Erodida_Dilatada (Opening)',imagem_cerebro_erodida_dilatada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Opening: Erosao e depois Dilataçao - comando Opening
imagem_cerebro_opening=cv2.morphologyEx(imagem_cerebro.copy(),cv2.MORPH_OPEN,None)
cv2.imshow('Imagem Original',imagem_cerebro)
cv2.imshow('Imagem Opening',imagem_cerebro_opening)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Closing: Dilataçao e depois Erosão (Tira o ruido preto - Pepper)
imagem_cerebro_dilatada=cv2.dilate(imagem_cerebro.copy(),(5,5),iterations=5)
imagem_cerebro_dilatada_erodida=cv2.erode(imagem_cerebro_dilatada.copy(),(5,5),iterations=5)
cv2.imshow('Imagem Original',imagem_cerebro)
cv2.imshow('Imagem Dilatada_Erodida (Closing)',imagem_cerebro_dilatada_erodida)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Closing: Dilataçao e depois Erosão (Tira o ruido preto - Pepper)
imagem_cerebro_closing=cv2.morphologyEx(imagem_cerebro.copy(),cv2.MORPH_CLOSE,None)
cv2.imshow('Imagem Original',imagem_cerebro)
cv2.imshow('Imagem Closing',imagem_cerebro_closing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Detecção de bordas de objetos na imagem

```python
#Canny Edge Detector
imagem_GoldenGate=cv2.imread('C:/Users/Falcone/Documents/0_MachineLearning/2022.07.08_Envirom_OpenCV/dicas-de-las-vegas-golden-gate_baixaresol.jpg') 
imagem_GoldenGateCinza=cv2.cvtColor(imagem_GoldenGate,cv2.COLOR_BGR2GRAY)
imagem_Golden_filtrada=cv2.GaussianBlur(imagem_GoldenGateCinza,(5,5),0)
#Calculando os filtros Canny com difierentes limiares
largo=cv2.Canny(imagem_Golden_filtrada,50,220) #limiar inferior, limiar superior
medio=cv2.Canny(imagem_Golden_filtrada,70,140)
apertado=cv2.Canny(imagem_Golden_filtrada,210,220)
#Mostrando as imagens
cv2.imshow('Imagem Original',imagem_Golden_filtrada)
cv2.imshow('Imagem largo',largo)
cv2.imshow('Imagem medio',medio)
cv2.imshow('Imagem apertado',apertado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Mostrando as bordas na imagem sem filtro
largo=cv2.Canny(imagem_GoldenGateCinza,50,220) #limiar inferior, limiar superior
medio=cv2.Canny(imagem_GoldenGateCinza,70,140)
apertado=cv2.Canny(imagem_GoldenGateCinza,210,220)
#Mostrando as imagens
cv2.imshow('Imagem Original',imagem_GoldenGateCinza)
cv2.imshow('Imagem largo',largo)
cv2.imshow('Imagem medio',medio)
cv2.imshow('Imagem apertado',apertado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
