```python
dados1=dados.drop(['age','Medu','Fedu','traveltime','studytime','failures','famrel','freetime','goout','Dalc',
                    'Walc','health','absences','G1','G2','G3'], axis=1)
```

```python
dados2=pd.get_dummies(dados1)
```

```python
dados3=dados[['age','Medu','Fedu','traveltime','studytime','failures','famrel','freetime','goout','Dalc',
                    'Walc','health','absences','G1','G2','G3']]
```

```python
#dados3
```

```python
dados4=pd.concat([dados2,dados3], axis=1)
```

```python
dados4
```

```python
dados4.dtypes
```

```python
dadosx=dados4.drop('G3', axis=1)
```

```python
dadosy=dados4['G3']
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
modelo=AdaBoostRegressor()
divdados=KFold(n_splits=3, shuffle=True)
resultado=cross_val_score(modelo,dadosx,dadosy,cv=divdados)
```
