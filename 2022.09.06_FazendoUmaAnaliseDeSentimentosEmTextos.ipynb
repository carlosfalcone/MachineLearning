```python
#filtrando somente as colunas de interesse
arquivo=arquivo[['text','stars']]
```

```python
#classificando ratings a partir de 4 estrelas como positivo e abaixo de 4 estrelas como negativo
arquivo['sentimento']=['positivo' if (x>3) else 'negativo' for x in arquivo['stars']]
```

```python
#passando todas as palavras para lower case (minusculas)
arquivo['text']=[x.lower() for x in arquivo['text']]
```

```python
#vendo o resultado e observando que existem caracteres especiais
pd.set_option('display.max_colwidth',None) #Mostrando toda a largura da coluna, sem abreviar
arquivo
```

```python
#eliminação dos caracteres especiais
import re
arquivo['text']=arquivo['text'].apply((lambda x: re.sub('[^a-z0-9\s]',' ',x))) #substituir os caracteres especiais por espaços
#pois algumas palavras podem se juntar e aí passa a nao ter mais sentido. Ex: thenit happenedshe (then it happened she)
```

```python
arquivo
```

```python
arquivo['text']=arquivo['text'].apply((lambda x: re.sub('[\n]',' ',x))) #substituir os caracteres especiais por espaços
```

```python
arquivo
```

```python
#tokenizando as palavras mais frequentes
#words=3000 #INPUT - tokenizando x palavras
from keras.preprocessing.text import Tokenizer
tokenizador=Tokenizer(num_words=words) #Quantidade de palavras a serem tokenizadas. Este codigo esta realmente limitando a quantidade de palavras tokenizadas? pois na celula abaixo todas elas foram tokenizadas!
tokenizador.fit_on_texts(arquivo['text'].values)

#Explicação sobre o comando num_words
# https://stackoverflow.com/questions/46202519/keras-tokenizer-num-words-doesnt-seem-to-work
#>>> t = Tokenizer(num_words=3)
#>>> l = ["Hello, World! This is so&#$ fantastic!", "There is no other world like this one"]
#>>> t.fit_on_texts(l)
#>>> t.word_index
#{'fantastic': 6, 'like': 10, 'no': 8, 'this': 2, 'is': 3, 'there': 7, 'one': 11, 'other': 9, 'so': 5, 'world': 1, 'hello': 4}

#There is nothing wrong in what you are doing. word_index is computed the same way no matter how many most frequent words 
#you will use later (as you may see here). So when you will call any transformative method - Tokenizer will use only three
#most common words and at the same time, it will keep the counter of all words - even when it's obvious that it will not 
#use it later.
```

```python
# Continuando a explicaçao acima
#Limiting num_words to a small number (eg, 3) has no effect on fit_on_texts outputs such as word_index, word_counts, 
#word_docs. It does have effect on texts_to_matrix. The resulting matrix will have num_words (3) columns.
array_token=tokenizador.texts_to_matrix(arquivo['text'], mode='count')
array_token.shape
```

```python
#Conforme explicado acima o vocabulario, enumeraçao das palavras nao muda, e é feito em todas elas
vocabulario=len((tokenizador.word_index))+1
vocabulario
```

```python
#o tokenizador foi criado, mas ate aqui o nosso dataset ainda nao mudou. Iremos criar uma variavel X com as palavras tokenizadas
X = tokenizador.texts_to_sequences(arquivo['text'].values)
```

```python
#mostrando as palavras da linha com indice 2
sel_arquivo=arquivo['text'].loc[2]
print(sel_arquivo.split()) #lista de palavras da linha indice 2 da tabela
print(len(sel_arquivo.split())) #quantidade de palavras da linha acima
```

```python
#verificando o tamanho do texto mais longo
max_comprim=max(len(i) for i in X)
print(max_comprim)
```

```python
#reescrevendo o codigo acima para poder encontrar a linha que possui essa quantidade de palavras
palavras=[]
comprim2=[]
contador=0
lista_contador=[]
guarda_valor=0
for i in X:
    palavras.append(i)
    compr=len(i) 
    if compr>guarda_valor:
        #print('guarda valor:',guarda_valor)
        guarda_valor=compr
        lista_contador.append(contador)
    contador=contador+1 
print('linha com o maximo de palavras:',max(lista_contador))
print('indice da linha:',len(X[max(lista_contador)]))
print('Lista de palavras:',X[max(lista_contador)])
```

```python
#criando lotes de dados temporais, fazendo com q cada amostra (linha com varias palavras) tenha o mesmo tamanho
#from keras.preprocessing.sequence import pad_sequences #nao funciona mais nesta versao de python
from keras_preprocessing.sequence import pad_sequences
X = pad_sequences(X)
```

```python
X[2]
```

```python
len(X[1])
```

```python
from sklearn.model_selection import train_test_split
Y=pd.get_dummies(arquivo['sentimento']).values #aplicando one-hot-encoding na variavel target
```

```python
Y
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(X,Y, test_size=0.3)
print('Tamanho do treino:',x_treino.shape, y_treino.shape)
print('Tamanho do teste:',x_teste.shape, y_teste.shape)
#teste 3000 linhas(amostras) com 974 palavras cada uma e variavel target com 3000 linhas e 2 tipos de classificaçao (positivo ou negativo)
```

```python
X.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM
import warnings
warnings.filterwarnings('ignore')
#criando a rede neural temporal
modelo=Sequential()
#modelo.add(Embedding(input_dim=vocabulario,output_dim=200,input_length=X.shape[1])) #criação de 200 caracteristicas para o vocabulario de palavras
modelo.add(Embedding(input_dim=words,output_dim=200,input_length=X.shape[1])) #criação de 200 caracteristicas para o vocabulario de palavras
modelo.add(LSTM(150,dropout=0.3)) #criando a LSTM (100 neuronios na camada oculta)
modelo.add(Dense(2,activation='softmax'))#definindo a saida com 2 classes
modelo.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#treinando o modelo:
modelo.fit(x_treino,y_treino,batch_size=500,epochs=epocas,verbose=2)
```

```python
#testando a performance do modelo nos dados de teste
__, acuracia= modelo.evaluate(x_teste,y_teste,verbose=2,batch_size=500)
print('Acuracia com dados de teste: %.2f' % (acuracia))
```

```python
# from keras.models import Sequential
# from keras.layers import Dense, Embedding, LSTM
# import warnings
# warnings.filterwarnings('ignore')
#Criando a rede neural temporal
modelo=Sequential()
modelo.add(Embedding(input_dim=modelo_word2vec.vectors.shape[0], output_dim=modelo_word2vec.vectors.shape[1],
                     weights=[modelo_word2vec.vectors],input_length=X.shape[1])) #criando os vetores work embedding pre-treinados
modelo.add(LSTM(150, dropout=0.3)) #criando a LSTM, com 150 neuronios na camada oculta
modelo.add(Dense(2,activation='softmax')) #definindo a saida com 2 classes (comentario positivo ou negativo)
modelo.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#Treinando o modelo
modelo.fit(x_treino,y_treino,batch_size=500, epochs=epocas, verbose=2)
```

```python
#testando a performance do modelo nos dados de teste
__, acuracia= modelo.evaluate(x_teste,y_teste,verbose=2,batch_size=500)
print('Acuracia com dados de teste: %.2f' % (acuracia))
```
