```python
import re
#passando todas as palavras para lowercase
lista_musicas=[x.lower() for x in lista_musicas]
```

```python
#Tirando os caracteres especiais
lista_musicas=[re.sub('[^a-z0-9\s]','',x) for x in lista_musicas]
```

```python
lista_musicas
```

```python
import pandas as pd
musicas_pandas=pd.DataFrame(lista_musicas,columns=['letras'])
```

```python
musicas_pandas.head()
```

```python
#Agrupando todas as frases - .cat() concatenate
texto_total = musicas_pandas['letras'].str.cat(sep=' ')
```

```python
texto_total
```

```python
texto_total.split()
```

```python
#eliminando possiveis espaços multiplos
#separa cada palavra do texto e depois junta as palavras inserindo um espaço entre elas
texto_total=' '.join(texto_total.split())
```

```python
texto_total
```

```python
texto_total=texto_total.split()
#para fazer isso, nao precisava ter juntado o texto acima! O Natanael fez acima um comando de juntar o texto q nao precisava
```

```python
texto_total
```

```python
len(texto_total)
```

```python
#agrupando sequencias de 5 palavras
tamanho_frases=5
sequencias_texto=[]
for i in range(tamanho_frases, len(texto_total)):
    sequencia=texto_total[i-tamanho_frases:i] #pega 4 palavras e prevê a 5ª
    sequencias_texto.append(sequencia)
```

```python
sequencias_texto
```

```python
import numpy as np
np.shape(sequencias_texto)
```

```python
#criando a variavel X com as palavras tokenizadas:
from keras.preprocessing.text import Tokenizer
tokenizador=Tokenizer()
tokenizador.fit_on_texts(sequencias_texto)
X = tokenizador.texts_to_sequences(sequencias_texto)
```

```python
X
```

```python
tamanho_vocabulario = len(tokenizador.word_index)+1
print('Tamanho do vocabulario de palavras:',tamanho_vocabulario)
```

```python
dataset=pd.DataFrame(X)
```

```python
dataset
```

```python
#criando as variaveis preditoras e variavel target
y=dataset[dataset.columns[4]]
x=dataset.drop(dataset.columns[4],axis=1)
```

```python
x
```

```python
y
```

```python
#transformando a variavel target em categorias
from keras.utils import to_categorical
y=to_categorical(y, num_classes=tamanho_vocabulario)
```

```python
dataframe_y=pd.DataFrame(y)
```

```python
dataframe_y.loc[:,25:42] 
```

```python
#Separando os dados em treino e teste
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste=train_test_split(x,y,test_size=0.3)
print('Tamanho do treino:',x_treino.shape, y_treino.shape)
print('Tamanho do teste:',x_teste.shape, y_teste.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM
import warnings
warnings.filterwarnings('ignore')
#criando a rede neural temporal
modelo=Sequential()
modelo.add(Embedding(input_dim=tamanho_vocabulario,output_dim=50,input_length=x_treino.shape[1])) #criação de 200 caracteristicas para o vocabulario de palavras
modelo.add(LSTM(100,dropout=0.2)) #criando a LSTM (100 neuronios na camada oculta)
modelo.add(Dense(tamanho_vocabulario,activation='softmax'))#definindo a saida
modelo.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#treinando o modelo:
modelo.fit(x_treino,y_treino,batch_size=500,epochs=200,verbose=2)
```

```python
#testando a performance do modelo nos dados de teste
__, acuracia= modelo.evaluate(x_teste,y_teste,verbose=2,batch_size=500)
print('Acuracia com dados de teste: %.2f' % (acuracia))
```
