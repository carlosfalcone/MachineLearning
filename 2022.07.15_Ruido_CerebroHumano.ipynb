```python
import cv2
```

```python
imagem=cv2.imread('C:/Users/Falcone/Documents/0_MachineLearning/2022.07.08_Envirom_OpenCV/Brain_SaltPepperNoise.jpg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(imagem)
```

```python
imagem.shape
```

```python
#Filtros utilizando o valor da mediana no kernel(filtro) AxA
filtro3=cv2.medianBlur(imagem,3)
filtro5=cv2.medianBlur(imagem,5)
filtro7=cv2.medianBlur(imagem,7)
cv2.imshow('Imagem Original',imagem)
cv2.imshow('Filtro 3x3',filtro3)
cv2.imshow('Filtro 5x5',filtro5)
cv2.imshow('Filtro 7x7',filtro7)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Filtro utilizando o valor da mdia
filtroAvg3=cv2.blur(imagem,(3,3))
filtroAvg7=cv2.blur(imagem,(7,7))
cv2.imshow('Imagem Original',imagem)
cv2.imshow('Filtro 5x5',filtroAvg3)
cv2.imshow('Filtro 7x7',filtroAvg7)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Filtro com media ponderada
filtroGau3=cv2.GaussianBlur(imagem,(3,3),0) #kernel 3x3 e desvio padrao 0
filtroGau5=cv2.GaussianBlur(imagem,(5,5),0)
filtroGau7=cv2.GaussianBlur(imagem,(7,7),0)
cv2.imshow('Imagem Original',imagem)
cv2.imshow('Filtro 3x3',filtroGau3)
cv2.imshow('Filtro 5x5',filtroGau5)
cv2.imshow('Filtro 7x7',filtroGau7)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
