```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagem=cv2.imread('C:/Users/Falcone/Documents/0_MachineLearning/2022.07.29_HaarCascade/duran-duran.jpg')
imagem2=cv2.cvtColor(imagem,cv2.COLOR_BGR2RGB)
#imagem_cinza=cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)
```

```python
classificador=cv2.CascadeClassifier('C:/Users/Falcone/Documents/0_MachineLearning/2022.07.29_HaarCascade/haarcascade_frontalface_alt.xml')
```

```python
retangulos_rostos=classificador.detectMultiScale(imagem2,1.1,5)
```

```python
for (x,y,largura,altura) in retangulos_rostos:
    imagem=cv2.rectangle(imagem2,(x,y),(x+largura,y+altura),(255,0,0),4)
plt.figure(figsize=(10,10))
plt.imshow(imagem2)
```
