```python
from sklearn.datasets import load_boston
import pandas as pd
dados=load_boston()
dadosx=pd.DataFrame(dados.data, columns=dados.feature_names)
dadosy=pd.Series(dados.target)
```

```python
from sklearn.model_selection import train_test_split
xtreino, xteste, ytreino, yteste = train_test_split(dadosx, dadosy, test_size=0.3)
```

```python
xtreino.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
modelo=Sequential()
modelo.add(Dense(20, input_dim=13, kernel_initializer='normal', activation='relu'))
modelo.add(Dense(20, kernel_initializer='normal', activation='relu'))
modelo.add(Dense(1, kernel_initializer='normal', activation='linear'))
```

```python
from keras.optimizers import Adam
otimizador=Adam()
```

```python
#modelo.compile(loss='mean_squared_error', optimizer=otimizador, metrics=['acc']) #Retorna valor zero para acc, pois esta
#metrica é utilizada somente em problemas de classificaçao
modelo.compile(loss='mean_squared_error', optimizer=otimizador, metrics=['mae'])
```

```python
historico=modelo.fit(xtreino, ytreino, epochs=1000, batch_size=354, validation_data=(xteste, yteste), verbose=1)
```

```python
import matplotlib.pyplot as plt
treino=historico.history['mae'] #mae
teste=historico.history['val_mae']
epochs=range(1,len(treino)+1)
plt.plot(epochs,treino,'-g', label='MAE Treino')
plt.plot(epochs,teste,'-b', label='MAE Teste')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('MAE')
```
