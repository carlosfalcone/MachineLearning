```python
dadosy=dados['mushroom']
```

```python
dadosy=dadosy.replace('EDIBLE',0)
dadosy=dadosy.replace('POISONOUS',1)
```

```python
dadosy.value_counts()
```

```python
dadosx=dados.drop('mushroom', axis=1)
```

```python
dadosx
```

```python
dadosx=pd.get_dummies(dadosx)
```

```python
dadosx
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
modelo=RandomForestClassifier(n_estimators=100,n_jobs=-1)
divdados=StratifiedKFold(n_splits=3,shuffle=True)
resultado=cross_val_score(modelo,dadosx,dadosy,cv=divdados,n_jobs=-1)
```
