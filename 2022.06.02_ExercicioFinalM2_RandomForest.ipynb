```python
dadosx=dados.drop('ClassMovment', axis=1)
```

```python
dadosy=dados['ClassMovment']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
modelo=RandomForestClassifier()
divdados=StratifiedKFold(n_splits=3)
resultado=cross_val_score(modelo,dadosx,dadosy,cv=divdados,n_jobs=-1)
```

```python
resultado.mean()
```

```python
from sklearn.model_selection import GridSearchCV
param={'n_estimators':[100,150,200,300],'min_samples_split':[2,3,4,5],'min_samples_leaf':[1,2,3,4]}
otimiza=GridSearchCV(modelo,param,cv=divdados,n_jobs=-1)
otimiza.fit(dadosx,dadosy)
print(otimiza.best_params_)
print(otimiza.best_score_)
#A otimizaÃ§Ã£o nao obteve ganho expressivo
```

```python
from sklearn.model_selection import cross_val_predict
predicao=cross_val_predict(modelo,dadosx,dadosy,cv=divdados,n_jobs=-1)
```
