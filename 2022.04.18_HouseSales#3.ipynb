```python
#2022.04.12 
#Fazer a regressao linear para varias variaveis, dividida entre dados de treino e teste
v2_dados_x = v1_dados.drop('price', axis = 1) 
v3_dados_y = v1_dados['price']
x_treino, x_teste, y_treino, y_teste = train_test_split(v2_dados_x, v3_dados_y, 
                                                        test_size = 0.30, random_state=14)
```

```python
#2022.04.18 
def modelosregressao(a, b, c, d):
    from sklearn.linear_model import LinearRegression
    from sklearn.linear_model import Ridge 
    from sklearn.linear_model import Lasso
    from sklearn.linear_model import ElasticNet
    #x_treino = a
    #y_treino = b
    #x_teste = c
    #y_teste = d
    reg = LinearRegression()
    ridge = Ridge()
    lasso = Lasso()
    elastic = ElasticNet()
    reg.fit(x_treino, y_treino)
    ridge.fit(x_treino, y_treino)
    lasso.fit(x_treino, y_treino)
    elastic.fit(x_treino, y_treino)
    ajuste_reg = reg.score(x_teste, y_teste)
    ajuste_ridge = ridge.score(x_teste, y_teste)
    ajuste_lasso = lasso.score(x_teste, y_teste)
    ajuste_elastic = elastic.score(x_teste, y_teste)
    print('Regressao Linear:', ajuste_reg, 'Ridge:', ajuste_ridge, 'Lasso:', ajuste_lasso, 'ElasticNet:', ajuste_elastic)
```

```python
modelosregressao(x_treino, y_treino, x_teste, y_teste)
```
